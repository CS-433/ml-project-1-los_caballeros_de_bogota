```python
import sys
import csv
import numpy as np
import matplotlib.pyplot as plt

sys.path.append('..')
from src.helpers import load_csv_data

DATA_PATH = '../data/train.csv'
```

```python
y, x, ids_train = load_csv_data(DATA_PATH, sub_sample=False)
```

```python
with open(DATA_PATH) as csv_file:

    csv_reader = csv.reader(csv_file, delimiter = ',')
    
    feature_names = []
    
    for row in csv_reader:
        feature_names.append(row)
        break

feature_names = feature_names[0][2:]
print("List of features names : ", feature_names)
```

```python
fig, axs = plt.subplots(nrows=6, ncols=5, figsize=(20,15))
fig.tight_layout(pad=1.5)
axs = axs.ravel()

for i, feature in enumerate(feature_names):
    mask = np.logical_not(x[:,i] == -999.0)
    axs[i].hist(x[mask,i], bins=100)
    axs[i].set_title(feature)
    if i in [0, 1, 2, 3, 5, 8, 9, 10, 13, 16, 19, 21, 23, 26, 29]:
        axs[i].set_yscale('log')
```
